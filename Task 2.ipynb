```python
# import libraries

import matplotlib.pyplot as mpl_pp
import pandas as pd
import geodatasets
import geopandas as gp
import numpy as np
```

```python
# Get the path to the dataset
# Load the dataset
# print output

world = gp.read_file(r"C:\Website\edgeUniPythonTest\DATA MINING AND VISUALISATION 2024\Exercise 2\110m_cultural\ne_110m_admin_0_countries.shp")
cities = gp.read_file(r"C:\Website\edgeUniPythonTest\DATA MINING AND VISUALISATION 2024\Exercise 2\110m_cultural\ne_110m_populated_places.shp")

print(world.columns.tolist())

print(cities.head())
```

```python
# Remove Antarctica
# Calculate GDP per capita
# Plot the Choropleth map
# print output

world = world[world['NAME'] != "Antarctica"]

world['gdp_per_capita'] = world['GDP_MD'] / world['POP_EST']

world.plot(column='gdp_per_capita', legend=True, figsize=(15, 10), cmap='Blues',
           legend_kwds={'label': "GDP per Capita", 'orientation': "horizontal"})
mpl_pp.title("GDP per Capita by Country")
mpl_pp.show()
```

```python
# Apply logarithmic scaling for better visualization
# Plot the Choropleth map with a diverging colormap

world['log_gdp_per_capita'] = np.log(world['gdp_per_capita'])

world.plot(column='log_gdp_per_capita', legend=True, figsize=(15, 10), cmap='RdYlGn',
           legend_kwds={'label': "Log GDP per Capita", 'orientation': "horizontal"})
mpl_pp.title("Global Inequality in GDP per Capita (Log Scale)")
mpl_pp.show()
```
